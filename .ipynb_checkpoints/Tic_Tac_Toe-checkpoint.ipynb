```python
player = 'X'
opponent = 'O'
```

```python
board = [
             [' ', ' ', ' '], 
             [' ', ' ', ' '], 
             [' ', ' ', ' ']
        ]
```

```python
def displayBoard(board):
    
    print("\n")
    print("\t" + "c-1" + " " + "c-2" + " " + "c-3")
    print("\n")
    print("r-1 " + "\t " + board[0][0] + " | " + board[0][1] + " | " + board[0][2])
    print("\t" + "---+---+---")
    print("r-2 " + "\t " + board[1][0] + " | " + board[1][1] + " | " + board[1][2])
    print("\t" + "---+---+---")
    print("r-3 " + "\t " + board[2][0] + " | " + board[2][1] + " | " + board[2][2])
    print("\n")
```

```python
def checkWinningCondition(board):
    
    #checking each rows for victory
    for row in range(3):
        if(board[row][0] == board[row][1] and board[row][1] == board[row][2]):
            if(board[row][0] == player):
                return 10
            elif(board[row][0] == opponent):
                return -10
    
    #checking each col for victory
    for col in range(3):
        if(board[0][col] == board[1][col] and board[1][col] == board[2][col]):
            if(board[0][col] == player):
                return 10
            elif(board[0][col] == opponent):
                return -10
            
    #checking each diagonal for victory
    
    #Diagonal-1
    if(board[0][0] == board[1][1] and board[1][1] == board[2][2]):
        if(board[0][0] == player):
            return 10
        elif(board[0][0] == opponent):
            return -10
            
    #Diagonal-2
    if(board[0][2] == board[1][1] and board[1][1] == board[2][0]):
        if(board[0][2] == player):
            return 10
        elif(board[0][2] == opponent):
            return -10
        
    return 0
```

```python
def gameOver(board):
    
    for row in range(3):
        for col in range(3):
            if(board[row][col] == " "):
                return False
    return True
```

```python
def checkInputValidity(row, col, board):
    
    if(row <= 0 or row >= 4):
        return False
    if(col <= 0 or col >= 4):
        return False
    if(board[row-1][col-1] != ' '):
        return False
    
    return True
```

```python
def takeUserInput(board):
    
    row = int(input("Please enter a valid the row number (1, 2, 3): "))
    col = int(input("Please enter a valid the column number (1, 2, 3): "))
    
    valid = checkInputValidity(row, col, board)
    
    while(not valid):
        print("Sorry! the input was invalid please enter a valid input...")
        
        row = int(input("Please enter a valid the row number (1, 2, 3): "))
        col = int(input("Please enter a valid the column number (1, 2, 3): "))
    
        valid = checkInputValidity(row, col, board)    
    
    board[row-1][col-1] = opponent
```

```python
def miniMax(board, depth, isMaxTurn):
    
    score = checkWinningCondition(board)
    
    #if the computer has won or lost return the score
    if(score == 10 or score == -10):
        return score
    
    #if no moves left, then it is atie
    if(gameOver(board)):
        return 0
    
    #if it is max's turn i.e isMaxTurn == True
    if(isMaxTurn):
        bestVal = -1000
        
        for row in range(3):
            for col in range(3):
                
                if board[row][col] == ' ':
                    
                    board[row][col] = player
                    bestVal = max(bestVal, miniMax(board, depth+1, not isMaxTurn))
                    
                    board[row][col] = ' '
                    
        return bestVal
    
    else:
        
        bestVal = 1000
        
        for row in range(3):
            for col in range(3):
                
                if board[row][col] == ' ':
                    
                    board[row][col] = opponent
                    bestVal = min(bestVal, miniMax(board, depth+1, not isMaxTurn))
                    
                    board[row][col] = ' '
                    
        return bestVal
```

```python
def bestPossibleMove(board):
    
    bestVal = -1000
    bestMove = (-1, -1)
    
    #checking the minimax value for each cell
    
    for row in range(3):
        for col in range(3):
            
            #if a position is empty
            if(board[row][col] == ' '):
                
                #make the move and calculate its minimax value
                board[row][col] = player
                moveVal = miniMax(board, 0, False)
                
                #undo the move
                board[row][col] = ' '
                
                #get the best move for the computer
                if(moveVal > bestVal):
                    bestMove = (row, col)
                    bestVal = moveVal
                    
    return bestMove
```

```python
def takeComputerInput(board):
    
    row, col = bestPossibleMove(board)
    board[row][col] = player   
```

```python
def playGame(board):
    
    print("Staring the Game...")
    print("You will play as O and the computer will play as X")
    
    turnCounter = 0
    over = gameOver(board)
    
    while(not over):
        
        won = checkWinningCondition(board)
        
        if(won == 10):
            print("X won !!! Better luck next time...")
            break
            
        elif(won == -10):
            print("O won !!! Congratulations...")
            break
        
        displayBoard(board)
        
        if(turnCounter%2 == 0):
            takeUserInput(board)
        else:
            takeComputerInput(board)
        
        turnCounter += 1
        over = gameOver(board)
        
    displayBoard(board)
```

```python
playGame(board)
```
